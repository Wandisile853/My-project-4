```python
plt.hist(wandi['Wear Specs']) 
plt.title('Histogram of Wear Specs') 
plt.xlabel('Values')
plt.ylabel('Frequency') 
plt.show() 
```

```python
sns.boxplot(x='Wear Specs', y='Education Status', data=wandi)
plt.title('Box Plot') 
plt.show() 
```

```python
x=np.arange (1,10,0.1)
y=2*x+5
plt.plot(x,y,color='r')
plt.title('line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show() 
```

```python
x=np.arange(1,10,0.1) 
y1=2*x+5
y2=3*x+10
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.subplot(1,2,2)
plt.plot(x,y2) 
plt.show() 
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[1,2,3,4,5,4,3,2,9]
b=[5,6,7,8,3,2,1,4,9]
plt.scatter(x,a)
plt.scatter(x,b) 
plt.legend(['a','b']) 
plt.title('X vs a&b') 
plt.xlabel('x-axis')
plt.ylabel('y-axis') 
plt.show() 
```

```python
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]
data=list([one,two,three]) 
plt.violinplot(data,showmeans=True,showmedians=True)
plt.grid(True) 
plt.title('Distribution of Data')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
